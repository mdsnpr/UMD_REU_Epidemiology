```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def sir_model(y, t, pars):
    S = y[0]
    I = y[1]
    
    dSdt = -pars['beta'] * S * I
    dIdt = pars['beta'] * S * I - pars['gamma'] * I
    dRdt = pars['gamma'] * I
    dydt = [dSdt, dIdt, dRdt]
    return dydt

def sir_model(y, t, pars):
    S = y[0]
    I = y[1]
    
    dSdt = -pars['beta'] * S * I - pars['v'] * S
    dIdt = pars['beta'] * S * I - pars['gamma'] * I
    dRdt = pars['gamma'] * I + pars['v'] * S
    dydt = [dSdt, dIdt, dRdt]
    return dydt

def plotter(beta, gamma, N, I0, tf, st, seed):
    pars={}
    # pars['c'] = c # Contacts per unit time (days)
    # pars['p'] = p # Probability of infectious contact
    pars['beta'] = beta # Transmission rate 
    # pars['gamma'] = 1.0/TI # Recovery rate (1/days)
    pars['gamma'] = gamma # Recovery rate (1/days)
    pars['basR0'] = pars['beta'] / pars['gamma']
    pars['N'] = N
    pars['I0'] = I0
    pars['v'] = 0.05
    pars['S0'] = pars['N'] - pars['I0']
    
    # Deterministic Simulation
    t = np.arange(0, tf)
    y = integrate.odeint(sir_model, np.array([pars['S0'], pars['I0'], 0]) / pars['N'], t, args = (pars,))

    # Stochastic Simulation
    [tst, yst] = stochsim_SIR([pars['S0'], pars['I0'], 0], [0, tf], pars,seed)
    
    # Plot the results
    plt.plot(t, y[:, 0], label='Susceptible_DT', color='r')
    plt.plot(t, y[:, 1], label='Infectious_DT', color='g')
    plt.plot(t, y[:, 2], label='Recovered_DT', color='b')

    if st:
        plt.plot(tst, yst[:, 0]/pars['N'], label='Susceptible_ST', linestyle = '--', color='r')
        plt.plot(tst, yst[:, 1]/pars['N'], label='Infectious_ST', linestyle = '--', color='g')
        plt.plot(tst, yst[:, 2]/pars['N'], label='Recovered_ST', linestyle = '--', color='b')

        print('Initial Infectious Populations: \n')
        print(yst[:, 1][0:9])

    R_tf = y[:, 2][tf-1]

    dSdt_vals = y[:, 1] * y[:, 0] * beta
    total_infection = np.trapz(dSdt_vals, t)
            
    I_peak = np.max(y[:, 1])
    t_peak = t[np.argmax(y[:, 1])]
    print('\nPeak of Infectious Population: ' + str(I_peak) + ' at Time: ' + str(t_peak))
    print('Total Amount of Infection (proportion) - R_tf (approximately R_inf): ' + str(R_tf), ' OR Sum of -DS/DTs: ' + str(total_infection))
    
    plt.plot(t_peak, I_peak, 'ro')
    plt.text(t_peak + 1, I_peak, f'Peak\nDay {t_peak}\n{I_peak:.2%}', color='red')
    plt.title('Beta = ' + str(pars['beta']) + ', Gamma = ' + str(pars['gamma']))
    plt.xlabel( 'Time (days) ')
    plt.ylabel('Population fraction')
    plt.legend()
    plt.show()
    # print(y[:, 0])
    # print(y[:, 1])
    # print(y[:, 2])
```

```python
def stochsim_SIR(y0, trange, pars, seed):
    np.random.seed(seed)
    t0 = trange [0]
    tf = trange [1]
    t = [t0]
    # We start y as a list of [S, I,R], where each triplet represents
    # a time point
    y = [y0]
    tcur = t0
    ycur = list(y0)
    ind = 0
    # Model
    while tcur<tf:
    # Check to see if there is an infection
        if ycur[1]==0:
            ind = ind+1
            t.append(tf)
            y.append(ycur)
            break
            
        # Rates
        infrate = pars['beta']*ycur[0]*ycur[1]/pars['N']
        recrate = pars['gamma']*ycur[1]
        
        totrate = infrate + recrate
        dt = -1/totrate * np.log(np.random.uniform())
        tcur = tcur + dt
        # Event type
        if np.random.uniform() < (infrate/totrate) : # Infection
            ycur[0] = ycur[0] - 1
            ycur[1] = ycur[1] + 1
        else: # Recovery
            ycur[1] = ycur[1] - 1
            ycur[2] = ycur[2] + 1
        ind = ind + 1
        t.append(tcur) 
        y.append(list(ycur))
    # Now recast y as a numpy array to make it easier to work with
    y = np.array(y)
    return [t, y]
```

# Different Types of Output in Stochastic Simulation

## Fizzle

when disease failed to spread in the beginning of the outbreak.

possible small fluctations in the beginning OR immediate extinction of disease

- Fizzle case tends to **decrease** when
    - Increase beta (transmission rate)
    - Decrease gamma (recovery rate)
    - Increase initial infectious population

while fixing any other parameters

## Burnout

when I(t) reaches 0 after an outbreak.

- Burnout case tends to **increase** when
    - with same modifications above

- Burnout case may **decrease** if
    - susceptible influx is added to the model
        - prevents susceptible depletion --> vulnerable for reignition of disease

## Persistent

when I(t) continues to be > 0 for a long period of time

- Potential for second outbreak
- Hard to observe in this basic model without Susceptible Influx via Vital Dynamics, Waning Immunity, etc.
     - with these there may be always have a chance for disease to reignite by preventing susceptible depletion
